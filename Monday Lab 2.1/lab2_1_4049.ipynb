```python
import statistics

def analyze_numbers(numbers):
  """
  Calculates and prints the mean, minimum, and maximum of a list of numbers.

  Args:
    numbers: A list of numerical values.
  """
  if not numbers:
    print("The list is empty. Cannot calculate statistics.")
    return

  mean_value = statistics.mean(numbers)
  min_value = min(numbers)
  max_value = max(numbers)

  print(f"--- Number Analysis Results ---")
  print(f"Input List: {numbers}")
  print(f"Mean: {mean_value:.2f}")
  print(f"Minimum: {min_value}")
  print(f"Maximum: {max_value}")
  print(f"-------------------------------")
```

```python
# Run the function with the provided list
my_list = [12, 7, 9, 21, 5]
analyze_numbers(my_list)
```

```python
def is_armstrong(number):
  """
  Checks if a given number is an Armstrong number.

  An Armstrong number is a number that is equal to the sum of its own digits
  each raised to the power of the number of digits.

  Args:
    number: An integer.

  Returns:
    True if the number is an Armstrong number, False otherwise.
  """
  num_str = str(number)
  num_digits = len(num_str)
  sum_of_powers = 0
  for digit_char in num_str:
    digit = int(digit_char)
    sum_of_powers += digit ** num_digits

  if sum_of_powers == number:
    print(f"{number} is an Armstrong number.")
    return True
  else:
    print(f"{number} is not an Armstrong number.")
    return False
```

```python
# Test the function with 153 and 123
is_armstrong(153)
is_armstrong(123)
```
