```python
# Task 1
import statistics

def analyze_numbers(numbers):
    """
    Calculates and prints the mean, minimum, and maximum of a list of numbers.

    Args:
        numbers (list): A list of numerical values.
    """
    if not numbers:
        print("The list is empty. Cannot calculate statistics.")
        return

    mean_value = statistics.mean(numbers)
    min_value = min(numbers)
    max_value = max(numbers)

    print(f"List of numbers: {numbers}")
    print(f"Mean: {mean_value}")
    print(f"Minimum: {min_value}")
    print(f"Maximum: {max_value}")

# Run the function on the specified list
my_list = [12, 7, 9, 21, 5]
analyze_numbers(my_list)
```

```python
# Weather dataset preprocessing
import pandas as pd
import numpy as np

# Load the dataset (update the path as needed)
data = pd.read_csv('weather_forecast_data.csv')

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Handle missing values in 'temperature' and 'humidity' (fill with median)
for col in ['temperature', 'humidity']:
    if col in data.columns:
        data[col].fillna(data[col].median(), inplace=True)

# Create new features: 'season' and 'week_number'
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

if 'date' in data.columns:
    data['season'] = data['date'].dt.month.apply(get_season)
    data['week_number'] = data['date'].dt.isocalendar().week

# Normalize 'rainfall' values (min-max scaling)
if 'rainfall' in data.columns:
    min_rain = data['rainfall'].min()
    max_rain = data['rainfall'].max()
    data['rainfall_normalized'] = (data['rainfall'] - min_rain) / (max_rain - min_rain)

# Show the processed dataframe
data.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer, MinMaxScaler

# Load the dataset
movies = pd.read_csv('IMDB-Movie-Data.csv')

# Remove duplicate movie entries based on Title and Year
movies = movies.drop_duplicates(subset=['Title', 'Year'])

# Handle missing values in 'Rating' and 'Genre'
if 'Rating' in movies.columns:
    movies['Rating'].fillna(movies['Rating'].median(), inplace=True)
if 'Genre' in movies.columns:
    movies['Genre'].fillna('Unknown', inplace=True)

# Encode 'Genre' as multi-hot (one-hot for each genre)
if 'Genre' in movies.columns:
    mlb = MultiLabelBinarizer()
    genre_lists = movies['Genre'].str.split(',')
    genre_encoded = mlb.fit_transform(genre_lists)
    genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)
    movies = pd.concat([movies, genre_df], axis=1)

# Encode 'Language' if present
if 'Language' in movies.columns:
    movies['Language'] = movies['Language'].astype(str)
    movies['Language_encoded'] = movies['Language'].astype('category').cat.codes

# Scale 'Rating' between 0 and 1
if 'Rating' in movies.columns:
    scaler = MinMaxScaler()
    movies['Rating_scaled'] = scaler.fit_transform(movies[['Rating']])

# Save the cleaned and processed dataframe to a new CSV
movies.to_csv('IMDB-Movie-Data-processed.csv', index=False)

# Show the first few rows
print(movies.head())


# Explanation in simple words:
# 1. We load the movie dataset from a CSV file.
# 2. We remove duplicate movie entries based on Title and Year.
# 3. We handle missing values in 'Rating' and 'Genre'.
# 4. We encode 'Genre' as multi-hot (one-hot for each genre).
# 5. We encode 'Language' as numeric codes.
# 6. We scale 'Rating' to be between 0 and 1 using MinMaxScaler.
```
